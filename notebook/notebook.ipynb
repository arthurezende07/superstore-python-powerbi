Tratamento e Limpeza de Dados

```python
#realizando alterações em cópia do dataframe
df = df_base.copy()
```

```python
##renomeando colunas
df.columns = df.columns.str.lower()
df.columns = df.columns.str.lower().str.replace(' ', '_')
df.columns = df.columns.str.lower().str.replace('-', '_')
df.columns = df.columns = df.columns.str.strip()
df.columns
```

```python
#removendo coluna row_id
df = df.drop('row_id', axis = 1)
df.head()
```

```python
#verificando tipos de dados
df.info()
```

```python
#converter datas para datetime
for col in ["order_date", "ship_date"]:
    df[col] = pd.to_datetime(df[col], format="%m/%d/%Y")
```

```python
# Foi verificada duplicidade considerando order_id e product_id.
# Embora existam múltiplas ocorrências dessa combinação, a maioria apresenta diferenças em métricas como discount, quantity ou profit, indicando que representam transações distintas dentro do mesmo pedido.
# Apenas uma duplicata exata foi identificada
```

```python
#removendo duplicatas
df = df.drop_duplicates()
df.info()
```

```python
#padronizando nome des de colunas
cols_texto = ['customer_name', 'city', 'state'] 
for col in cols_texto: 
    df[col] = df[col].str.strip().str.title()
df.head()
```

Criação de Colunas Derivadas

```python
#prazo de entrega - período do pedido até a data do envio
df['data_entrega'] = (df['ship_date'] - df['order_date']).dt.days
```

```python
#componentes temporais com relação a data do pedido
df['ano'] = df['order_date'].dt.year
df['mes'] = df['order_date'].dt.month
df['trimestre'] = df['order_date'].dt.quarter
```

```python
#margem de lucro por venda
df['margem_pct'] = (df['profit']/df['sales']).round(4)
```

```python
#ticket médio por pedido
ticket = df.groupby('order_id')['sales'].sum().reset_index()
ticket.columns = ['order_id', 'ticket_medio']
df = df.merge(ticket, on ='order_id', how='left')
```
